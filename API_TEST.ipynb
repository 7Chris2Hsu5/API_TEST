```python
import requests
import json
from bs4 import BeautifulSoup
from tkinter import *
```

```python
def toJson(data):
    data = eval(data)
    data = json.dumps(data)    
    return data


def GET(url, output):
    response = requests.get(url)
    
    if response.ok:
        try:
            rj = response.json()
            if isinstance(rj, list):
                s = ''
                for i in range(len(rj)):
                    s += str(rj[i]) + '\n'
                output['text'] = s
            else:
                output['text'] = rj
        except:
            output['text'] = response.content
    else:
        output['text'] = response
       
    
def POST(url, body, output):
    body_json = toJson(body)
    response = requests.post(url, data=body_json, headers={'content-type': 'application/json'})
    
    if response.ok:
        output['text'] = response.text
    else:
        output['text'] = response

        
def PUT(url, body, output):
    body_json = toJson(body)
    response = requests.put(url, data=body_json, headers={'content-type': 'application/json'})

    if response.ok:
        output['text'] = response.text
    else:
        output['text'] = response


def DELETE(url, output):
    response = requests.delete(url)
    
    if response.ok:
        output['text'] = response.text
    else:
        output['text'] = response
```

```python
# http://localhost:8080/student
if __name__ == '__main__':
    
    window = Tk()
    window.title("Call API")
    window.geometry("850x550")
    
    frame = Frame(window)
    frame.pack()
    
    bottomframe = Frame(window)
    bottomframe.pack()

    url_label = Label(frame, text='API', font=("標楷體", 16)).pack()
    
    url = Entry(frame, bd=2, width=60)
    url.pack()
    
    body_label = Label(frame, text='BODY(json)', font=("標楷體", 16)).pack()
    
    body = Text(frame, width=60, height=15)
    body.pack()
    
    output = Label(window, text='', font=("標楷體", 16))

    GET_Button = Button(frame, text="GET", font=('Saab', 18), command=lambda: GET(url.get(), output))
    GET_Button.pack(side=LEFT, expand=True, fill=X)
    
    POST_Button = Button(frame, text="POST", font=('Saab', 18), command=lambda: POST(url.get(), body.get('1.0', END), output))
    POST_Button.pack(side=LEFT, expand=True, fill=X)
    
    PUT_Button = Button(frame, text="PUT", font=('Saab', 18), command=lambda: PUT(url.get(), body.get('1.0', END), output))
    PUT_Button.pack(side=LEFT, expand=True, fill=X)
    
    DELETE_Button = Button(frame, text="DELETE", font=('Saab', 18), command=lambda: DELETE(url.get(), output))
    DELETE_Button.pack(side=LEFT, expand=True, fill=X)
    
    output.pack()

    window.mainloop()
```
